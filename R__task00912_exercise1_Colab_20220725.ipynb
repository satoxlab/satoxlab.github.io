## 46行目解説
```
 1. baculoviralProteinIDs <-

 3.                                 baculoviridaeFasta@ranges@NAMES, 
 4.                                                                  function(x) 
 5.                                                                              str_split(x, pattern=" ")
 6.                                                                                                       [[1]]

```
Rの関数実行には順番がある。この例においてsapply (simgplified (l)apply)を使った場合には、  
3.のbaculoviridaeFasta@ranges@NAMESの要素が順番に4-7の処理に用いられる。  
4.では3から受け取っbaculoviridaeFasta@ranges@NAMESの中身の1つを5.にxとして渡し5.のstr_split関数で空白（pattern=" ")ごとにxを分割する。その結果は「リスト」形式で出力される。  
6.の[[1]]は「1番目のリスト」を指定し、7.ではstr_splitで出力された各リストの実体「ベクトル」の1番目の要素を指定している。  
  sapplyはbaculoviridaeFasta@ranges@NAMESの要素全てについて高速に4-7の計算を行い、それ出力全てを1.で指定しているbaculoviralProteinIDsオブジェクトとして格納する。

# get line indexes to parse BLAST output scores

````r
queryLineIDs               <- grep("^Query= ", blastOutputLines, perl=TRUE)
hitInformationStartLineIDs <- grep("^Sequences producing significant alignments:", blastOutputLines, perl=TRUE)
alignmentStartLineIDs      <- grep("^ALIGNMENTS$", blastOutputLines, perl=TRUE)
````

## 解説: 

grep（"パターン", 検索対象文字列）でパターンで指定した文字列が検索対象文字列に含まれるか検索。

perl=TRUEでプログラミング言語Perlで用いられているパターン指定の文法を利用する。"^文字列"で行の先頭に文字列がある場合を指定。



## check length of parsed IDs to see they are matched.

## define functions: 関数の定義

````r
get.query.id <- function(blastOutputLine){
    strsplit(blastOutputLine, "(Query= | )", perl=TRUE)[[1]][2]
}

get.query.id

get.scores <- function(blastOutputLine){
    splitResult <- strsplit(blastOutputLine, "\\s+", perl=TRUE)[[1]]
    bitScore    <- as.numeric(splitResult[length(splitResult)-1])
    evalue      <- as.numeric(splitResult[length(splitResult)])
    return(list("id"=splitResult[1], "bitscore"=bitScore, "evalue"=evalue))
}

get.scores
````

## 解説: Rのデータの種類

> numeric: 数値  
> integer: 整数  
> character: 文字  
> factor: 因子

````r
queryIDs  <- sapply(queryLineIDs, function(x) get.query.id(blastOutputLines[x]))
scoreList <- sapply(1:length(hitInformationStartLineIDs), 
                    function(x) 
                      matrix(sapply(blastOutputLines[(hitInformationStartLineIDs[x]+2):(alignmentStartLineIDs[x]-2)], 
                                    get.scores), 
                             ncol=3, byrow=TRUE))
names(scoreList) <- queryIDs

sapply(scoreList, dim) # check scoreList by examing object dimensions
````

##===========================================================================##
## EXERCISE1.8: 行列データを作る、ループ制御を使う
##===========================================================================##

````r
## define function
get.species.name <- function(x){
    # x: blastOutputTable[RI,1]
    baculoviralProteinIndex <- match(x, baculoviralProteinIDs) #matchでbaculoviralProteinIDsのxを含むものを抜き出す
    nudiviralProteinIndex   <- match(x, nudiviralProteinIDs)

    if (is.na(baculoviralProteinIndex) == FALSE){
        return(baculoviralNames[names(baculoviralProteinIDs[baculoviralProteinIndex])])
    }else{
        if (is.na(nudiviralProteinIndex) == FALSE){
            return(nudiviralNames[names(nudiviralProteinIDs[nudiviralProteinIndex])])
        }else{
            ## optional: stop due to not viral species names were not found
        }
    }
}

blastMatchMatrix <- matrix(NA, 
                           nrow=length(bmnpvProteinIDs), 
                           ncol=length(unique(nudiviralNames))+length(unique(baculoviralNames)), 
                           dimnames=list(bmnpvProteinIDs, 
                                         c(sort(unique(as.character(baculoviralNames))), 
                                           sort(unique(as.character(nudiviralNames))))))
homologMatrix <- matrix(NA, 
                        nrow=length(bmnpvProteinIDs), 
                        ncol=length(unique(nudiviralNames))+length(unique(baculoviralNames)), 
                        dimnames=list(bmnpvProteinIDs, 
                                      c(sort(unique(as.character(baculoviralNames))), 
                                        sort(unique(as.character(nudiviralNames))))))

for (bmnpvProteinID in as.character(bmnpvProteinIDs)){
    blastOutputTable <- scoreList[[bmnpvProteinID]]
    
    for (RI in 1:nrow(blastOutputTable)){
        speciesName <- as.character(get.species.name(blastOutputTable[RI,1]))
        homologMatrix[bmnpvProteinID, speciesName] <- 
            as.character(ifelse (is.na(blastMatchMatrix[bmnpvProteinID, speciesName]) == TRUE,
                    blastOutputTable[RI,1],
                    ifelse(blastMatchMatrix[bmnpvProteinID, speciesName] > blastOutputTable[RI,3],
                           blastOutputTable[RI,1], homologMatrix[bmnpvProteinID, speciesName])))    # 2017-05-11 修正箇所
        blastMatchMatrix[bmnpvProteinID, speciesName] <-                                            # 以前のコードだと1種のウイルスから
            as.numeric(ifelse (is.na(blastMatchMatrix[bmnpvProteinID, speciesName]) == TRUE,        # 複数個のBLASTマッチがある場合に
                    blastOutputTable[RI,3],                                                         # 謝ってNAが入ってしまう
                    ifelse(blastMatchMatrix[bmnpvProteinID, speciesName] > blastOutputTable[RI,3],
                           blastOutputTable[RI,3], blastMatchMatrix[bmnpvProteinID, speciesName]))) # 2017-05-11 修正箇所
    }

}

str(blastMatchMatrix)
str(homologMatrix)
````

##===========================================================================##
## EXERCISE1.9: ループ制御を使って行列からデータを抽出し、可視化する
##===========================================================================##

# filter out possible false positives:
# BLASTの計算で確率的に偶然ヒットしてしまったかもしれない偽陽性ヒット候補を除外する

BLASTに用いた検索対象タンパク質数（データベースサイズ）を求める

````r
databaseSize <- length(baculoviridaeFasta) + length(nudiviridaeFasta)
databaseSize
````

各遺伝子のBLASTマッチのE-valueをヒストグラムで閾値とともに可視化する。

````r
for (RI in 1:nrow(blastMatchMatrix)){
    if (length(na.omit(blastMatchMatrix[RI,])) == 1) next
    rangeValues <- summary(na.omit(blastMatchMatrix[RI,]))[c("Min.", "Max.")]
    baculoHistData <- hist(blastMatchMatrix[RI,1:73], breaks=seq(rangeValues[1], rangeValues[2], by=(rangeValues[2]-rangeValues[1])/100), plot=FALSE)
    nudiHistData <- hist(blastMatchMatrix[RI,74:79], breaks=seq(rangeValues[1], rangeValues[2], by=(rangeValues[2]-rangeValues[1])/100), plot=FALSE)

    png(paste(rownames(blastMatchMatrix)[RI], "_e-value.png", sep=""), width=480*4, height=480*3, res=300)
    plot(NA, xlim=rangeValues, ylim=c(0, max(c(max(baculoHistData$counts), max(nudiHistData$counts)))), main="", xlab="E-value", ylab="Frequency")    
    invisible(sapply(1:100, function(x) rect(xleft=baculoHistData$breaks[x], xright=baculoHistData$breaks[x+1], ybottom=0, ytop=baculoHistData$counts[x])))
    invisible(sapply(1:100, function(x) rect(xleft=nudiHistData$breaks[x], xright=nudiHistData$breaks[x+1], ybottom=0, ytop=nudiHistData$counts[x], col="magenta")))
    abline(v=1/databaseSize, col="magenta", lty=2)
    dev.off()
}
````

##===========================================================================##
## EXERCISE1.10: ループ制御を使って行列データを修正し、得られた遺伝子ファミリーの各タンパク質配列をファイルへの書き出す。
##===========================================================================##

閾値以上のE-valueを持つマッチは除外し、残った遺伝子を遺伝子ファミリーとしてfastaファイルを書き出す。

また、遺伝子ファミリー中のバキュロウイルス属、ヌディウイルス属のタンパク質数を出力する。


````r
filteredHomologMatrix <- homologMatrix

for (RI in 1:nrow(homologMatrix)){
    filteredHomologMatrix[RI, which(blastMatchMatrix[RI,] > 1/databaseSize)] <- NA
    namesForFasta <- as.character(na.omit(filteredHomologMatrix[RI, ]))
    baculoviridaeHits <- sum(sapply(namesForFasta, function(x) length(grep(x, baculoviridaeFasta@ranges@NAMES))))
    nudiviridaeHits   <- sum(sapply(namesForFasta, function(x) length(grep(x, nudiviridaeFasta@ranges@NAMES))))
    cat(paste(rownames(homologMatrix)[RI], paste(baculoviridaeHits, nudiviridaeHits, sep="\t"), sep=": ")) 
    cat("\n")
    fastaIDs <- sapply(namesForFasta, function(x) grep(x, allFasta@ranges@NAMES))
    writeXStringSet(allFasta[fastaIDs], filepath=paste(rownames(homologMatrix)[RI], "_homologs_filtered.fasta", sep=""))
}
````
